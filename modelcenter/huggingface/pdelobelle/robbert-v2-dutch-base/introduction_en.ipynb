```python
import paddle
from paddlenlp.transformers import AutoModel

model = AutoModel.from_pretrained("pdelobelle/robbert-v2-dutch-base")
input_ids = paddle.randint(100, 200, shape=[1, 20])
print(model(input_ids))
```
